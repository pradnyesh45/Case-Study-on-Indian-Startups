```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
import csv
startup_funding = pd.read_csv('./startup_funding.csv')
startup_funding
```

```python
# Trying to solve 1st question using pandas
```

```python
df = startup_funding.copy()
df
```

```python
df1 = df.filter(['Date','AmountInUSD'], axis=1)
df1
```

```python
df1['Date'].isnull().sum()
```

```python
df1['AmountInUSD'].isnull().sum()
```

```python
df1.dropna(subset = ["AmountInUSD"], inplace=True)
df1
```

```python
df1['AmountInUSD'] = df1['AmountInUSD'].str.replace(',', '')
df1
```

```python
mapping = {}
for ind in df1:
    length = len(df1['Date'][ind])
    year = df1['Date'][ind][length - 4: ]
    if year in mapping:
        mapping[year] += 1
    else:
        mapping[year] = 1 
mapping
```

```python
mapping = collections.OrderedDict(sorted(mapping.items()))
mapping
```

```python
year = []
funding = []
for i in mapping:
    year.append(int(i))
    funding.append(mapping[i])
print(year)
print(funding)
```

```python
plt.plot(year, funding)
plt.xticks(rotation=45)
plt.show()
for i in mapping:
    print(i, mapping[i])
```
