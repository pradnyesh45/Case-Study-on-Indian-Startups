```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import collections
startup_funding = pd.read_csv('./startup_funding.csv')
startup_funding
```

```python
df = startup_funding.copy()
df
```

```python
df1 = df.filter(['Date','AmountInUSD'], axis=1)
df1
```

```python
df1['Date'].isnull().sum()
```

```python
df1['AmountInUSD'].isnull().sum()
```

```python
df1.dropna(subset = ["AmountInUSD"], inplace=True)
df1
```

```python
df1['AmountInUSD'] = df1['AmountInUSD'].str.replace(',', '')
df1
```

```python
mapping = {}
for ind in df1.index:
    length = len(df1['Date'][ind])
    if df1['Date'][ind] not in mapping:
        mapping[df1['Date'][ind][length - 4: length]] = 0
    mapping[df1['Date'][ind][length - 4: length]] += int(df1['AmountInUSD'][ind]) 
mapping
```

```python
mapping = collections.OrderedDict(sorted(mapping.items()))
mapping
```

```python
year = []
funding = []
for i in mapping:
    year.append(int(i))
    funding.append(mapping[i])
print(year)
print(funding)
```

```python
plt.plot(year, funding)
plt.show()
for i in mapping:
    print(i, mapping[i])
```
